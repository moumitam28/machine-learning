```python
import pandas as pd
```

```python
df = pd.read_csv("missing.csv", parse_dates =["day"]) # string ---> dates
df.set_index('day', inplace =True) # making the day col as index
```

```python
df
```

## Fillna

```python
new_df = df.fillna(0)
new_df
```

## We can fill different columns with different values too

```python
# we can do it by using dictionary
new_df = df.fillna({
    'temperature':0,
    'windspeed':0,
    'event': 'No event'
})
new_df
# but filling with '0' is not a good idea , so we can carry forward with previous data only
# in that case, 2017-01-04 : value will be same as 2017-01-01 and it is 32.0
```

```python
new_df =  df.fillna(method = 'ffill') # we can also fill with method = 'bfill' , in that case the NaN will be filled with next value
new_df
```

```python
new_df =  df.fillna(method = 'ffill', limit = 1) # copying value only one column down
new_df
```

## Interpolate

```python
#print(df)
new_df = df.interpolate(method = 'index') # dy default the interpolate method is 'linear' but we can use lot of functions like 'quadratic', 'index', 'values', 'time'
new_df
```

## Dropna

```python
new_df = df.dropna() # drop the rows which is having atleast one missing value
print(new_df)
# lets we want to delete where all the values in a row are not there
new_df =df.dropna(how="all")
new_df
```

```python
# we can also set how many NaN value we wanna have in that case we can use  'thresh' for 'THRESHOLD'
new_df = df.dropna(thresh = 2)
new_df
```
